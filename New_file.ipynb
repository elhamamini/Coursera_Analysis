```python
import pandas as pd
import json
from pandas import json_normalize
from sklearn.preprocessing import LabelEncoder
from ua_parser import user_agent_parser
from user_agents import parse
```

```python
df = pd.read_csv('new_txt.txt', sep=",",header=None,on_bad_lines='skip')
df
```

```python
result = df.to_json(orient="split")
```

```python
result
```

```python
parsed = json.loads(result)

List_of_lists=parsed["data"]
List_of_lists
```

```python
list_of_dict=[]
for line in List_of_lists:
    obj={}
  
    for row in line:
      
        command,description = row.strip().split(":",1)
         
        if command=="{key":
                command="key"
      
        obj[command]=description 
        
   
    list_of_dict.append(obj)
               
```

```python
#I did not used read bc first we have to create it
#with open("my_data.json", "r") as final:
    #json.dumps(list_of_dict, indent=4) 
```

```python
#"Writing mode to create a my_data.json file"
with open("my_data.json", "w") as final:
    json.dump(list_of_dict, final)
 
```
